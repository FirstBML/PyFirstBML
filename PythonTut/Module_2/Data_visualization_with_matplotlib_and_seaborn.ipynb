### Data visualization
### Data Visualization is the process of representing data graphically to reveal patterns, trends, correlations, and outliers that might not be obvious in raw numbers. It’s essential for data exploration, communication, and decision-making.

### Matplotlib:
#### A versatile Python library for creating a wide range of plots, from simple lines to complex 3D visuals. It gives you precise control over every element—colors, labels, sizes, and more—making it ideal for custom visualizations.

### Seaborn:
#### A higher-level library built on Matplotlib, designed for statistical data visualization. It simplifies creating complex plots, integrates easily with Pandas DataFrames, and offers attractive default styles and color palettes.

### Use Case:
#### Use these tools to analyze datasets like sales records, scientific measurements (e.g., flower dimensions in the Iris dataset), or survey results to identify trends or relationships.

### Matplotlib
#### Matplotlib is a powerful, flexible Python library for creating static, animated, and interactive visualizations. It’s widely used for data visualization because it allows you to generate a variety of plots—line plots, scatter plots, histograms, bar plots, box plots, and more—while giving you fine-grained control over every aspect of the figure.




### Install/Import Matplotlib

```python
import matplotlib.pyplot as plt
```

### Basic Settup

```python
Student = ['Jude', 'Ada', 'John', 'Grace', 'Bashiru', 'Samson']
Score = [55, 45, 60, 40, 70, 60]
```

### Line Plot

```python
plt.Figure(figsize = (10, 3))

plt.plot(Student, Score, color = 'blue')
plt.title('Students Test Scores')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.show()
```

### Scatter Plot

```python
plt.scatter(Student,Score, color ='green' )
plt.title('Students text scores')
plt.xlabel('Students name')
plt.ylabel('Score')
plt.show()
```

```python
plt.figure(figsize = (6, 3))

plt.bar(Student, Score, color = 'purple')
plt.title('Students text scores')
plt.xlabel('Students name')
plt.ylabel('Score')
plt.show()

```

### Histogram

```python
plt.figure(figsize=(4,4))


plt.hist(Score, color = 'black')
plt.title('Students text scores')
plt.xlabel('Students name')
plt.ylabel('Score')
plt.show()
```

### Seaborn
#### Seaborn is a super easy tool in Python that helps you draw pretty pictures (called plots) of your data. It’s like a magic crayon box built on top of Matplotlib, making graphs look nice and simple to create for beginners! Why Use It? It turns numbers into cool, colorful pictures so you can see patterns fast.



#### Install/Import Seaborn

```python
import seaborn as sns
import pandas as pd
```

### Basic Settup

```python
Student = ['Jude', 'Ada', 'John', 'Grace', 'Bashiru', 'Samson']
Score = [55, 45, 60, 40, 70, 60]

data = pd.DataFrame({'Student': Student, 'Score': Score})

data
```

### Line Plot

```python
sns.lineplot(x = Student,data=data, y= Score, color = 'blue')
plt.title('Students Score')
plt.xlabel('Students Name')
plt.ylabel('Score')
plt.show()

```

### Scatter Plot

```python
sns.scatterplot(x = Student,data=data, y= Score, color = 'purple')
plt.title('Students Score')
plt.xlabel('Students Name')
plt.ylabel('Score')
plt.show()

```

### Bar chart

```python
sns.barplot(x = Student, y= Score,data=data , color = 'green')
plt.title('Students Score')
plt.xlabel('Students Name')
plt.ylabel('Score')
plt.show()
```

### Histogram plot

```python
sns.histplot(x='Score', data=data, color = 'pink')
plt.title('How score Pile up')

plt.show()
```

```python
sns.set_style('white')  # this line is correct

sns.histplot(x='Score', data=data, color='pink', edgecolor=None, linewidth=0)  # fixed typo here

plt.title('How score Pile up')
plt.show()
```

### Visualizing Web3 Transfer Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\g\OneDrive\Documents\Python Problems\web3_token_transfers.csv', parse_dates= ['timestamp'])
df['date'] = df['timestamp'].dt.date

# set seaborn theme

sns.set(style='whitegrid')
```

```python
df.head()
```

```python
plt.Figure(figsize=(10,5))

sns.histplot(df['amount'], bins= 50, kde= True,color='Skyblue')
plt.title('Token Amount Transfer Distribution')
plt.xlabel('Amount')
plt.ylabel('frequency')
plt.xlim( 0,3000)
plt.show()
```

#### Daily Total transfer

```python
Daily_Trans = df.groupby('date')['amount'].sum().reset_index()

plt.figure(figsize=(10,6))

sns.lineplot(data=Daily_Trans, x='date', y='amount', marker='o')
plt.title('Daily Total Amount Transfer')
plt.ylabel('Amount')
plt.xlabel('Date')
plt.tight_layout()
plt.show()

```
