```python
# EXERCISE 1 — Even Numbers (Rewrite)
# [1, 2, 3, 4, 5, 6]

def compute_even_number(number):
    return [num for num in number if num % 2 == 0]    
print(compute_even_number([1, 2, 3, 4, 5, 6]))

```

```python
# EXERCISE 2 — Win/Loss Classification (Rewrite)
# Input: [10, -5, 7, -2]

def PnL_classification(profits):
    return ['Win' if p > 0 else 'Loss' for p in profits]
print(PnL_classification([10, -5, 7, -2]))
```

```python
# PART 2 — DICTIONARY PATTERNS (VERY IMPORTANT)
# EXERCISE 3 — Aggregate PnL Per Trader

""" 
Input: 
trades = [
 {"trader": "A", "pnl": 100},
 {"trader": "B", "pnl": -50},
 {"trader": "A", "pnl": 70},
 {"trader": "B", "pnl": 30}
]

"""
def aggregate_trader_pnl(trades):
    
    results = {}
    
    for trade in trades:
        Trader = trade['trader']
        PnL = trade['pnl']
        
        if Trader in results:
            results[Trader] += PnL
        else:
            results[Trader] = PnL
    
    return results

trades = [
 {"trader": "A", "pnl": 100},
 {"trader": "B", "pnl": -50},
 {"trader": "A", "pnl": 70},
 {"trader": "B", "pnl": 30}
]

print(aggregate_trader_pnl(trades))
```

```python
# PART 2 — DICTIONARY PATTERNS (VERY IMPORTANT)
# EXERCISE 4 — Count trades Per Trader

def Aggregate_Trader_count(trades):
    
    counts = {}
    
    for trade in trades:
        trader = trade['trader']
                
        if trader in counts:
            counts[trader] += 1
        else:
            counts[trader] = 1
    return counts

print(Aggregate_Trader_count(trades))
```

```python
def trade_count(trads):
    counts = {}
    for trade in trades:
        trader = trade['trader']
        counts[trader] = counts.get(trader, 0) + 1
    return counts

print(trade_count(trades))
```

```python
# PART 3 — STRING & RECORD PARSING
#EXERCISE 5 — Parse Trade String
# Input: "BTC,100,110,2"

def parse_trade_string(trades_str):
    symbol, entry_price, exit_price, qty = trades_str.split(",")
    
    entry_price = float(entry_price)
    exit_price = float(exit_price)
    qty = float(qty)
    
    PnL = (exit_price-entry_price)*qty
    
    return {"symbol":symbol,
              "entry_price":entry_price,
              "exit_price":exit_price,
              "quantity":qty,
              "Profit":PnL
              }

print(parse_trade_string("BTC, 100, 110, 2.5"))
```

```python
# PART 5 — CUMULATIVE METRICS (DATA SCIENCE CORE)
# EXERCISE 7 — Running Average

def cummulative_average(metrics):
    
    running_average = []
    total = 0
           
    for i, metric in enumerate(metrics, start = 1):
        total = metric + total
        average = total/i
        running_average.append(average)
        
    return running_average
        
print(cummulative_average([10, 20, 30, 40]))
```

```python
# PART 5 — CUMULATIVE METRICS (DATA SCIENCE CORE)
# EXERCISE 7 — Running Average
metrics = [10, 20, 30, 40]
def cummulative_average(metrics):
    
    running_average = []
    running_total = []
    total = 0
           
    for i, metric in enumerate(metrics, start = 1):
        total += metric
        average = total / i
        running_average.append(average)
        running_total.append(total)
               
    return running_average, running_total       
        
running_average, running_total = cummulative_average(metrics)

print(f' Running total: {running_total}')
print(f' Runing average: {running_average}')
```

```python
# PART 6 — ADVANCED 
# EXERCISE 8 — Top Trader by Total PnL

PnL_dict = {
 "A": 170,
 "B": -20,
 "C": 300
}


def top1_trader(PnL_dict):
    best_trader=None
    best_pnl=float('-inf')
    
    for trader, pnl in PnL_dict.items():
        if pnl > best_pnl:
            best_pnl = pnl
            best_trader = trader
    return best_trader, best_pnl

best_trader, best_pnl = top1_trader(PnL_dict)

print(f'Best trader is {best_trader} with profit of ${best_pnl}')
```
