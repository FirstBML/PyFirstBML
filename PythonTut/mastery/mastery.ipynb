```python
# EXERCISE 1: Sum of Even Numbers

def sum_of_even_numbers(numbers):
    total = 0
    
    for num in numbers:
        if num % 2 == 0:
            total += num
            
    return total

print(sum_of_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
# EXERCISE 2: Count Frequency of Items
def count_frequency(items):
    freq = {}
    
    for item in items:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1
            
    return freq

print(count_frequency(["BTC", "ETH", "BTC", "SOL", "ETH", "BTC"]))

```

```python
# exercise 3: Find Maximum Value

def find_max(numbers):
    max_val = numbers[0]
    
    for num in numbers:
        if num > max_val:
            max_val = num
            
    return max_val

print(find_max([5, 2, 9, 1, 7]))
```

```python
# EXERCISE 4 â€” Remove Duplicates
# define the variable carrying the set of numbers
def remove_duplicates(nums):
    seen = set()
    result = []
    
    for num in nums:
        if num not in seen:
            seen.add(num)
            result.append(num)
            
    return result

print(remove_duplicates([1, 2, 2, 3, 1, 4]))
```
