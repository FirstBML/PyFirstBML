```python
# sum even numbers
def even_number(nums):
    total = 0
    for num in nums:
        if num % 2 == 0:
            total += num
    return total        
print(even_number([1,2,3,4,5,6,7]))
```

```python
# EXERCISE 2: Count Frequency

def frequent_list(items):
    freqz = {}
    for item in items: # search through all the list in the items
        if item in freqz: # loof for item in freqz list
            freqz[item] += 1 # add 1 to the  exist count of that item if found 
        else:
            freqz[item] = 1 # if not found, add  just 1
    return freqz
print(frequent_list(["BTC", "ETH", "BTC", "SOL", "ETH", "BTC"]))

          
```

```python
# EXERCISE 3: Find Maximum Value [5, 2, 9, 1, 7]

# define the function
def find_max(number):
    #set max number to start at 0
    max_val = number[0]
    # search for numbers i list of numbers: for loop
    for num in number:
        # if the search number is greater than the first max number, 
        if num > max_val:
            # make max number the searched number
            max_val = num
    # retrun the max number after all searches
    return max_val
print(find_max ([5, 2, 9, 1, 7]))
    
```

```python
# EXERCISE 4: Remove Duplicates:[1, 2, 2, 3, 1, 4]

def remove_duplicate(number):
    seen = set()
    result = []
    for num in number:
        if num not in seen:
            seen.add(num)
            result.append(num)
    return result
print(remove_duplicate([1, 2, 2, 3, 1, 4]))
```

```python
# EXERCISE 5 — Basic PnL Calculator
# Input: entry=100, exit=110, qty=2

def Calculate_PnL(entry_price,exit_price,qty):
    entry_vol=entry_price*qty
    exit_vol=exit_price*qty
    PnL=exit_vol-entry_vol
    return PnL
print(Calculate_PnL(100,110,2))
```

```python
# EXERCISE 6 — Win/Loss Classification
# Input: [10, -5, 7, -2]
def Classifications(numbers):
    class_results = [] # create an empty list to accomodate the output
    for num in numbers:
        if num > 0:
            class_results.append('win')
        else:
            class_results.append('loss')
    return class_results
print(Classifications([10, -5, 7, -2]))
```

```python
# EXERCISE 7 — Filter Whale Trades
# Input:  {"value": 50000},  {"value": 120000},  {"value": 300000}]

def whales_trades(trades):
    whales = []
    for trade in trades:
        if trade['value'] > 100000:
            whales.append(trade)
    return whales
print(whales_trades([{"value": 50000},  {"value": 120000},  {"value": 300000}]))
```

```python
# EXERCISE 8 — Cumulative PnL
# Return cumulative PnL over time
# Input: [10, -5, 7]

def cummulative_PnL(profits):
    results = []
    cumPnL = 0
    for profit in profits:
        cumPnL += profit
        results.append(cumPnL)
    return results
print(cummulative_PnL([10, -5, 7]))
```

```python
# Write a function that returns the sum of all even numbers in a list.
# Input:  [1, 2, 3, 20, 4, 5, 6]

def even_number_sum(number):
    sum = 0
    for num in number:
        if num % 2 == 0:
            sum += num
    return sum
print(even_number_sum([1, 2, 3, 20, 4, 5, 6]))
```

```python
# count frequency of occurence
# ["BTC", "ETH", "BTC", "SOL", "ETH", "BTC"]

def frequency_count(items):
    results = {}
    for item in items:
         if item in results:
            results[item] += 1
         else:
            results[item] = 1
    return results
print(frequency_count(["BTC", "ETH", "BTC", "SOL", "ETH", "BTC"]))

```

```python
# Find Maximum Value [5, 2, 9, 1, 7]
def maximum_value(number):
    max_value = number[0]
    for num in number:
        if num > max_value:
            max_value = num
    return max_value
print(maximum_value([5, 2, 9, 1, 7]))
```
